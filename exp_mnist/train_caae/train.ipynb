```python
from XAE.sampler import gaus, multinomial
from XAE.util import calculate_sharpness, calculate_sharpness_generator
```

```python
test_generator = torch.utils.data.DataLoader(model.test_data, 16, num_workers = 5, shuffle = False, pin_memory=True, drop_last=True)
test_sample_image, test_sample_label = next(iter(test_generator))
recon_sample = model(test_sample_image.to(device) , test_sample_label.to(device)).to('cpu').detach().numpy()
```

```python
test_sample = np.array(test_sample_image)
ghx = []
for i in range(2):
    ghx.append(np.concatenate([test_sample[8*i+j,0,:,:] for j in range(8)], axis = 1))
for i in range(2):
    ghx.append(np.concatenate([recon_sample[8*i+j,0,:,:] for j in range(8)], axis = 1))
recon_results = np.concatenate(ghx, axis = 0)
```

```python
plt.figure(figsize = (10.0, 10.0))
plt.imshow(recon_results, cmap='gray', vmin=-1, vmax=1)
plt.show()
```

```python
test_z = []
test_label = []
z = gaus(6, 8)

for i in range(10):
    test_z.append(z)
    test_label.append(model.test_data.code[i].repeat(6,1))
    
test_z = torch.cat(test_z, dim = 0)
test_label = torch.cat(test_label, dim = 0)

Z = torch.cat((test_z, test_label), dim = 1).to(device)
```

```python
gen_sample = model.dec(Z).to('cpu').detach().numpy() # generated samples in numpy

gz = []
for i in range(10):
    gz.append(np.concatenate([gen_sample[6*i + j,0,:,:] for j in range(6)], axis = 1))
generation_result = np.concatenate(gz, axis = 0) # concatenate matrix for plt
```

```python
plt.figure(figsize = (6.0, 10.0))
plt.imshow(generation_result, cmap='gray', vmin=-1, vmax=1)
plt.show()
```

```python
z_sampler = lambda x : torch.cat([gaus(x,8), multinomial(x,10)], dim = 1).to(device)
```

```python
calculate_sharpness_generator(model.dec, z_sampler, batch_size=100, repeat = 10, device=device).mean()
```

```python
fid_model = fid.InceptionV3().to(device)
```

```python
m1, s1 = fid.calculate_activation_statistics(test_data2, fid_model, batch_size=100, dims=2048, device = device)
```

```python
act = fid.get_activations_generator(model.dec, z_sampler, fid_model, batch_size=100, repeat = 10, dims=2048, device=device)
m2 = np.mean(act,axis = 0)
s2 = np.cov(act, rowvar=False)
```

```python
fid.calculate_frechet_distance(m1, s1, m2, s2)
```
